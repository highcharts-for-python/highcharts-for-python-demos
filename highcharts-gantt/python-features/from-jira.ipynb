# ``.from_jira()`` Demo
This demonstrates the ``.from_jira()`` method which generates a Gantt chart from an Asana project.

## Import Dependencies

```python
import os
from highcharts_gantt.chart import Chart
from dotenv import load_dotenv

## Load Environment Variables
load_dotenv()
```

## Assemble Chart

```python
kwargs = {}
kwargs['project_key'] = os.getenv('JIRA_PROJECT_ID')

chart = Chart.from_jira(**kwargs)
```

## Display Chart

```python
chart.display()
```
